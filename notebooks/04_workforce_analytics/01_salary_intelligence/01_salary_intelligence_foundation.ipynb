### Analysis 1 - Salary by Experience Level

This analysis evaluates how salary changes as professionals progress through different experience levels.

```python
# Calculate descriptive salary statistics across experience levels

salary_by_experience_level = (
    df.groupby("experience_level")
      .agg(
          job_posting_count=("salary_usd", "count"),
          average_salary=("salary_usd", "mean"),
          median_salary=("salary_usd", "median"),
          minimum_salary=("salary_usd", "min"),
          maximum_salary=("salary_usd", "max"),
          salary_standard_deviation=("salary_usd", "std")
      )
      .round(2)
      .sort_values(by="average_salary", ascending=False)
)

salary_by_experience_level
```

## Workforce Salary Progression

Ranking is useful not only across workforce groups but also across career progression.

## Workforce Ranking

Dense ranking assigns the same rank to equal values without leaving gaps in the ranking sequence.

This provides an alternative perspective on workforce performance.

```python
# Rank job roles based on average salary

average_salary_by_job_role = (
    df.groupby("job_role")["salary_usd"]
      .mean()
)

job_role_salary_rank = (
    average_salary_by_job_role
    .rank(
        ascending=False,
        method="dense"
    )
    .astype(int)
    .sort_values()
)

job_role_salary_rank
```

```python
# Rank countries based on average salary

country_salary_rank = (
    df.groupby("country")["salary_usd"]
      .mean()
      .sort_values(ascending=False)
      .rank(
          ascending=False,
          method="dense"
      )
      .head(5)
)

country_salary_rank
```

## Salary Distribution Analysis

The following analyses summarize salary distribution using percentiles and quartiles, providing a deeper understanding of compensation across the AI workforce.

```python
# Calculate salary percentiles

salary_percentile_values = np.percentile(
    df["salary_usd"],
    [10, 30, 50, 75, 100]
)

salary_percentile_summary = pd.DataFrame({
    "Percentile": [
        "10%",
        "30%",
        "50%",
        "75%",
        "100%"
    ],
    "Salary": salary_percentile_values
})

salary_percentile_summary
```

```python
# Calculate bonus percentiles

bonus_percentile_values = np.percentile(
    df["bonus_usd"],
    [10, 30, 50, 75, 100]
)

bonus_percentile_summary = pd.DataFrame({
    "Percentile": [
        "10%",
        "30%",
        "50%",
        "75%",
        "100%"
    ],
    "Bonus": bonus_percentile_values,
    "Country": country_salary_rank
})

bonus_percentile_summary
```

```python
# Create salary quartile categories

df["salary_quartile"] = pd.qcut(
    df["salary_usd"],
    q=4,
    labels=[
        "Low",
        "Mid",
        "High",
        "Best"
    ]
)

df["salary_percentile"].value_counts(
    normalize=True
).mul(100)
```

## Compensation Spread

Spread measures the distance between the highest and lowest values within a distribution.

These metrics provide a quick indication of compensation inequality across the workforce.

```python
# Calculate salary and bonus spread

salary_spread = (
    df["salary_usd"].max() - df["salary_usd"].min()
)

bonus_spread = (
    df["bonus_usd"].max() - df["bonus_usd"].min()
)

distribution_spread_summary = pd.DataFrame({
    "Metric": [
        "Salary Spread",
        "Bonus Spread"
    ],
    "Value": [
        salary_spread,
        bonus_spread
    ]
})

distribution_spread_summary
```

## Statistical Variability

Variance and standard deviation measure how widely salaries vary across the workforce.

Higher values indicate greater salary dispersion.

```python
# Calculate salary variance

salary_variance = np.var(
    df["salary_usd"]
).round(3)

salary_variance
```

```python
# Calculate salary standard deviation

salary_standard_deviation = np.std(
    df["salary_usd"]
).round(3)

salary_standard_deviation
```
