## Level 6: Cohort Analytics

### Objective

Cohort Analytics studies how compensation evolves throughout an AI professional's career.

Unlike descriptive analytics, cohort analytics focuses on progression across different career stages and identifies where the largest improvements in salary, bonus, and total compensation occur.

This section answers questions such as:

- How does salary evolve across experience levels?
- How does bonus change during career progression?
- How does total compensation grow?
- Which career stage experiences the highest compensation growth?
- How much does compensation increase from Entry level to Lead level?

### 1.1 Environment Setup
Configure the notebook execution environment and suppress non-critical runtime warnings.

```python
import warnings
import numpy as np
import pandas as pd

# Configure execution environment and suppress warnings
warnings.filterwarnings("ignore")

# Optimize pandas display configurations for comprehensive data inspection.
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", None)

# Log framework metadata and verification metrics.

print("Environment Information")
print("-" * 22)
print(f"Pandas Version : {pd.__version__}")
print(f"NumPy Version  : {np.__version__}")
```

### 1.2 Data Ingestion
Load the core processed Global AI Jobs dataset.

```python
# Load target dataset

DATA_PATH = "../../../data/processed/cleaned_ai_jobs.csv"
df = pd.read_csv(DATA_PATH)
print("\nDataset Loaded Successfully")

# Execute descriptive matrix shape verification.

print("\nDataset Shape")
print("-" * 16)

total_rows, total_columns = df.shape

print(f"Rows    : {total_rows:,}")
print(f"Columns : {total_columns}")
print("-" * 16)

# Display a quick preview of the dataset schema.

display(df.head())
```

## 6.1 Experience Level Progression Analysis

Measure the average salary, bonus, and total compensation across experience levels, and calculate the percentage growth between consecutive career stages.

```python
# Define the experience level hierarchy

experience_level_order = [
    "Entry",
    "Mid",
    "Senior",
    "Lead"
]

# Calculate average salary by experience level

average_salary_by_experience = (
    df.groupby("experience_level")["salary_usd"]
      .mean()
      .reindex(experience_level_order)
      .round(2)
)

average_salary_by_experience
```

```python
# Calculate salary growth across experience levels

salary_growth_percentage = (
    average_salary_by_experience
    .pct_change()
    .mul(100)
    .round(2)
)

salary_growth_percentage
```

## 6.3 Analyze Bonus Progression Across Experience Levels

Evaluate how bonus compensation changes as professionals progress through different experience levels.

```python
# Calculate average bonus by experience level

average_bonus_by_experience = (
    df.groupby("experience_level")["bonus_usd"]
      .mean()
      .reindex(experience_level_order)
      .round(2)
)

average_bonus_by_experience
```

```python
# Calculate bonus growth across experience levels

bonus_growth_percentage = (
    average_bonus_by_experience
    .pct_change()
    .mul(100)
    .round(2)
)

bonus_growth_percentage
```

## 6.4 Analyze Total Compensation Progression Across Experience Levels

Measure the overall compensation growth throughout the career journey.

```python
# Calculate total compensation for every workforce record

df["total_compensation"] = (
    df["salary_usd"]
    + df["bonus_usd"]
)

df["total_compensation"].sample(5).reset_index()

# Calculate average total compensation by experience level

average_total_compensation_by_experience = (
    df.groupby("experience_level")["total_compensation"]
      .mean()
      .reindex(experience_level_order)
      .round(2)
)

average_total_compensation_by_experience
```

```python
# Calculate total compensation growth across experience levels

total_compensation_growth_percentage = (
    average_total_compensation_by_experience
    .pct_change()
    .mul(100)
    .round(2)
)

total_compensation_growth_percentage
```

## 6.5 Create Career Stage Categories

Segment job postings into career stages using years of professional experience.

```python
# Categorize job postings into career stages

df["career_stage"] = pd.cut(
    df["experience_years"],
    bins=[-1, 2, 5, 10, 20],
    labels=[
        "Early Career",
        "Developing",
        "Experienced",
        "Expert"
    ]
)

df["career_stage"].value_counts().sort_index()
```

## 6.6 Evaluate Salary Distribution by Career Stage

Analyze descriptive salary statistics for each career stage.

```python
# Calculate descriptive salary statistics by career stage

salary_by_career_stage = (
    df.groupby("career_stage")
      .agg(
          job_posting_count=("salary_usd", "count"),
          average_salary=("salary_usd", "mean"),
          median_salary=("salary_usd", "median"),
          minimum_salary=("salary_usd", "min"),
          maximum_salary=("salary_usd", "max"),
          salary_standard_deviation=("salary_usd", "std")
      )
      .round(2)
)

salary_by_career_stage
```

## 6.7 Evaluate Bonus Distribution by Career Stage

Analyze descriptive bonus statistics across career stages.

```python
# Calculate descriptive bonus statistics by career stage

bonus_by_career_stage = (
    df.groupby("career_stage")
      .agg(
          job_posting_count=("bonus_usd", "count"),
          average_bonus=("bonus_usd", "mean"),
          median_bonus=("bonus_usd", "median"),
          minimum_bonus=("bonus_usd", "min"),
          maximum_bonus=("bonus_usd", "max"),
          bonus_standard_deviation=("bonus_usd", "std")
      )
      .round(2)
)

bonus_by_career_stage
```

## 6.8 Evaluate Total Compensation Distribution by Career Stage

Analyze descriptive total compensation statistics across career stages.

```python
# Calculate descriptive total compensation statistics by career stage

compensation_by_career_stage = (
    df.groupby("career_stage")
      .agg(
          job_posting_count=("total_compensation", "count"),
          average_total_compensation=("total_compensation", "mean"),
          median_total_compensation=("total_compensation", "median"),
          minimum_total_compensation=("total_compensation", "min"),
          maximum_total_compensation=("total_compensation", "max"),
          total_compensation_standard_deviation=("total_compensation", "std")
      )
      .round(2)
)

compensation_by_career_stage
```

## 6.9 Measure Career Mobility

Calculate the percentage increase in salary, bonus, and total compensation from Entry level to Lead level.

```python
# Calculate salary mobility

salary_mobility_percentage = (
    (
        average_salary_by_experience["Lead"]
        - average_salary_by_experience["Entry"]
    )
    / average_salary_by_experience["Entry"]
) * 100

print(f"Salary Mobility: {salary_mobility_percentage:.2f}%")
```

```python
# Calculate bonus mobility

bonus_mobility_percentage = (
    (
        average_bonus_by_experience["Lead"]
        - average_bonus_by_experience["Entry"]
    )
    / average_bonus_by_experience["Entry"]
) * 100

print(f"Bonus Mobility: {bonus_mobility_percentage:.2f}%")
```

```python
# Calculate total compensation mobility

total_compensation_mobility_percentage = (
    (
        average_total_compensation_by_experience["Lead"]
        - average_total_compensation_by_experience["Entry"]
    )
    / average_total_compensation_by_experience["Entry"]
) * 100

print(
    f"Total Compensation Mobility: "
    f"{total_compensation_mobility_percentage:.2f}%"
)
```

## 6.10 Summarize Career Progression Metrics

Combine salary, bonus, and total compensation into a single summary table for comparative analysis across experience levels.

```python
# Create a consolidated career progression summary

career_progression_summary = pd.DataFrame({
    "Average Salary": average_salary_by_experience,
    "Average Bonus": average_bonus_by_experience,
    "Average Total Compensation": average_total_compensation_by_experience
})

career_progression_summary
```

## Level 8: Feature Engineering & Workforce Metrics

## Objective

Create meaningful workforce metrics by engineering new features from existing data. These features improve business analysis, support predictive modeling, and demonstrate practical feature engineering skills used in AI Engineering and Data Science.

## 8.1 Engineer Compensation-Based Metrics

Create new compensation metrics to better understand bonus contribution, compensation growth, and salary efficiency.

```python
# Create bonus percentage for each job posting

df["bonus_percentage"] = (
    (df["bonus_usd"] / df["salary_usd"]) * 100
).round(2)

# Review the distribution of bonus percentage

bonus_percentage_summary = (
    df["bonus_percentage"]
    .describe()
    .round(2)
)

bonus_percentage_summary
```

```python
# Create the total compensation multiplier

df["compensation_multiplier"] = (
    df["total_compensation"] / df["salary_usd"]
).round(3)

# Review the distribution of compensation multiplier

compensation_multiplier_summary = (
    df["compensation_multiplier"]
    .describe()
    .round(3)
)

compensation_multiplier_summary
```

```python
# Calculate salary earned per year of professional experience

df["salary_per_experience_year"] = np.where(
    df["experience_years"] == 0,
    df["salary_usd"],
    (df["salary_usd"] / df["experience_years"]).round(2)
)

# Review the distribution of salary per experience year

salary_per_experience_year_summary = (
    df["salary_per_experience_year"]
    .describe()
    .round(2)
)

salary_per_experience_year_summary
```

## 8.2 Engineer Salary Value Metrics

Develop workforce metrics that measure salary value relative to market demand and overall compensation ranking.

```python
# Create the salary demand ratio

df["salary_demand_ratio"] = (
    df["salary_usd"] /
    df["skill_demand_score"]
).round(2)

# Review the distribution of the salary demand ratio

salary_demand_ratio_summary = (
    df["salary_demand_ratio"]
    .describe()
    .round(2)
)

salary_demand_ratio_summary
```

```python
# Create the compensation score using percentile ranking

df["compensation_score"] = (
    (
        df["salary_usd"] +
        df["bonus_usd"]
    )
    .rank(pct=True)
    .round(4)
)

# Review the distribution of the compensation score

compensation_score_summary = (
    df["compensation_score"]
    .describe()
    .round(4)
)

compensation_score_summary
```
