```python
# Import Libraries

import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

# Configure Display Settings

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", None)

# Display Environment Information

print("\nEnvironment Information")
print("-" * 22)

print(f"Pandas Version : {pd.__version__}")
print(f"NumPy Version  : {np.__version__}")
```

```python
# Load Dataset (Global AI Jobs)

DATA_PATH = "../data/processed/cleaned_ai_jobs.csv"

df = pd.read_csv(DATA_PATH)

print("\nDataset Loaded Successfully")
```

```python
# Dataset Shape Analysis

print("\nDataset Shape")

print("-" * 16)

rows, columns = df.shape

print(f"Rows    : {rows:,}")
print(f"Columns : {columns}")

print("-" * 16)

# Dataset Preview

display(df.head())
```

```python
df.groupby("ai_specialization")["salary_usd"].max().round(2)
df.groupby("job_role")["salary_usd"].max().round(2)
df.groupby("country")["salary_usd"].max().sort_values(ascending=False)
df.groupby("hiring_difficulty_score")["salary_usd"].max().sort_values(ascending=False).head()
df.groupby("hiring_difficulty_score")["salary_usd"].min().sort_values(ascending=False).head()
df.groupby("experience_level")["salary_usd"].max().round(2).sort_values(ascending=False)
df.groupby("experience_years")["salary_usd"].max().round(2).sort_values()
df.groupby("industry")["salary_usd"].max().round(2).sort_values(ascending=False)
df.groupby("education_required")["salary_usd"].max().round(2).sort_values(ascending=False)
```

### Analysis 1
### Experience Level vs Salary

```python
salary_by_experience_level = (
    df.groupby(
        "experience_level"
    )["salary_usd"]
    .agg([
        "count",
        "mean",
        "median",
        "min",
        "max",
        "std"
    ])
    .round(2)
    .sort_values(
        by="mean",
        ascending=False
    )
)

salary_by_experience_level
```
