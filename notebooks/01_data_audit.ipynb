### Section 1: Environment Setup


```python
# Import Libraries

import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

# Configure Display Settings

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", None)

# Display Environment Information

print("\nEnvironment Information")
print("-" * 22)

print(f"Pandas Version : {pd.__version__}")
print(f"NumPy Version  : {np.__version__}")
```

### Section 3: Dataset Loading


```python
# Load Dataset (Global AI Jobs)

DATA_PATH = "../data/raw/ai_jobs.csv"

df = pd.read_csv(DATA_PATH)

print("\nDataset Loaded Successfully")
```

### Section 4: Dataset Snapshot


```python
# Dataset Shape Analysis

print("\nDataset Shape")

print("-" * 16)

rows, columns = df.shape

print(f"Rows    : {rows:,}")
print(f"Columns : {columns}")

print("-" * 16)

# Dataset Preview

display(df.head())
```

### Section 5: Structural Audit


```python
# Columns Audit

print("\n" + "-" * 20)
print("List of columns")
print("-" * 20 + "\n")

for index, col in enumerate(df.columns, start=1):
    print(f"{index}. {col}")
```

### Section 6: Data Type Analysis Audit


```python
print("-" * 20)
print("DATA TYPE ANALYSIS")
print("-" * 20)

datatype_summary = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.values
})

display(datatype_summary)

print("\nDatatype Counts:\n")
display(df.dtypes.value_counts())
```

### Section 7: Missing Value Audit


```python
# Missing Value Analysis

print("\n" + "-" * 22)
print("Missing Value Analysis")
print("-" * 22)

missing_values = df.isnull().sum()

missing_percentage = (
    df.isnull().sum() / len(df)
) * 100

missing_report = pd.DataFrame({
    "Missing Count": missing_values,
    "Missing Percentage": missing_percentage
})

display(
    missing_report.sort_values(
        by="Missing Count",
        ascending=False
    )
)
```

### Section 8: Empty String Validation

```python
print("-" * 23)
print("EMPTY STRING VALIDATION")
print("-" * 23)

categorical_columns = df.select_dtypes(
    include="object"
).columns

for col in categorical_columns:

    empty_count = (
        df[col]
        .astype(str)
        .str.strip()
        .eq("")
        .sum()
    )

    print(f"{col}: {empty_count}")
```

### Section 9: Score Validation

```python
# Score Validation

print("-" * 50)
print("Score Validation")
print("-" * 50)

numerical_columns = (
    df.select_dtypes(
        include=np.number
    ).columns
)

for col in numerical_columns:

    print(f"\n{col}")

    display (
        pd.DataFrame({
            "Min" : [df[col].min()],
            "Max" : [df[col].max()]
        })
    )
    # display(data)
```

### Section 9: Categorical Cardinality Analysis

```python
# Categorical Cardinality Analysis

print("\n" + "-" * 32)
print("Categorical Cardinality Analysis")
print("-" * 32)

cardinality_report = pd.DataFrame({
    "Column": categorical_columns,
    "Unique Values": [
        df[column].nunique()
        for column in categorical_columns
    ]
})

display(
    cardinality_report.sort_values(
        by="Unique Values",
        ascending=False
    )
)
```

### Section 10: Categorical Cardinality Analysis Review

```python
print("-" * 39)
print("Categorical Cardinality Analysis Review")
print("-" * 39)

categorical_columns = df.select_dtypes(
    include="object"
).columns

print(f"Total Categorical Columns: {len(categorical_columns)}")

for col in categorical_columns:
    display(
        df[col]
        .value_counts()
    )
```

### Section 11: Duplicates Analysis


```python
# Duplcate rows analysis

duplicate_count = df.duplicated().sum()

print(
    f"\nDuplicate Rows: {duplicate_count}"
)
```

### Section 9: Numerical Feature Analysis

```python
print("\n" + "-" * 26)
print("Numerical Feature Analysis")
print("-" * 26)

numerical_columns = (
    df.select_dtypes(
        include=np.number
    ).columns
)

print(f"Numerical Features : {len(numerical_columns)}")

display(
    df[numerical_columns]
    .describe()
    .T
    .round(2)
)
```

### Section 12: Numerical Validation Report

```python
# Numerical Validation Report

print("\n" + "-" * 27)
print("Numerical Validation Report")
print("-" * 27)

numerical_columns = (
    df.select_dtypes(
        include=np.number
        )
        .columns
    )

validation_report = []

for column in numerical_columns:

    validation_report.append({

        "Column"            :   column,
        "Minimum Value"     :   df[column].min(),
        "Maximum Value"     :   df[column].max(),
        "Negative Values"   :   (df[column] < 0).sum()

    })

validation_report_pd = pd.DataFrame(
    validation_report
)

display(validation_report_pd)
```

### Section 10: Outliers Investigation

```python
# Outliers Investigation

print("-" * 50)
print("Outliers Investigation")
print("-" * 50)

outlier_summary = []

for col in numerical_columns:

    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = (
        (
            df[col] < lower
        )
        |
        (
            df[col] > upper
        )
    ).sum()

    outlier_summary.append(
        [col, outliers]
    )

outlier_report = pd.DataFrame(
    outlier_summary,
    columns=[
        "Column",
        "Outlier Count"
    ]
)

display(
    outlier_report.sort_values(
        by="Outlier Count",
        ascending=False
    )
)
```

## Final Audit Summary


```python
audit_summary = pd.DataFrame({

    "Metric": [
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicate Rows",
        "Numerical Features",
        "Categorical Features"
    ],

    "Value": [
        len(df),
        len(df.columns),
        df.isnull().sum().sum(),
        df.duplicated().sum(),
        len(numerical_columns),
        len(categorical_columns)
    ]
})

print("\n" + "-" * 27)
print("Final Dataset Audit Summary")
print("-" * 27)

display(audit_summary)
```
