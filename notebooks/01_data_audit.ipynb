### Section 2: Environment Setup


```python
# Import Libraries

import pandas as pd
import numpy as np

# Configure Display Settings

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", None)

# Display Environment Information

print("\nEnvironment Information")
print("-" * 22)

print(f"Pandas Version : {pd.__version__}")
print(f"NumPy Version  : {np.__version__}")
```

### Section 3: Dataset Loading


```python
# Load Dataset (Global AI Jobs)

DATA_PATH = "../data/raw/ai_jobs.csv"

df = pd.read_csv(DATA_PATH)

print("\nDataset Loaded Successfully")
```

### Section 4: Dataset Snapshot


```python
# Dataset Shape Analysis

print("\nDataset Shape")

print("-" * 16)

rows, columns = df.shape

print(f"Rows    : {rows:,}")
print(f"Columns : {columns}")

print("-" * 16)

# Dataset Preview

display(df.head())
```

### Section 5: Structural Analysis


```python
# Columns

print("\n" + "-" * 20)
print("List of columns")
print("-" * 20 + "\n")

for index, column in enumerate(df.columns, start=1):
    print(f"{index}. {column}")
```

### Section 6: Data Type Audit


```python
print("\n" + '-' * 20)
print("Data Type Analysis")
print('-' * 20)

display(df.dtypes)

print("\nData Type Distribution")

display(df.dtypes.value_counts())
```

### Section 7: Missing Value Audit


```python
# Missing Value Analysis

print("\n" + "-" * 22)
print("Missing Value Analysis")
print("-" * 22)

missing_values = df.isnull().sum()

missing_percentage = (
    df.isnull().sum() / len(df)
) * 100

missing_report = pd.DataFrame({
    "Missing Count": missing_values,
    "Missing Percentage": missing_percentage
})

display(
    missing_report.sort_values(
        by="Missing Count",
        ascending=False
    )
)
```

### Section 8: Duplicate & Numerical Feature Audit


```python
# Duplcate rows analysis

duplicate_count = df.duplicated().sum()

print(
    f"\nDuplicate Rows: {duplicate_count}"
)
```

```python
print("\n" + "-" * 26)
print("Numerical Feature Analysis")
print("-" * 26)

numerical_columns = (
    df.select_dtypes(
        include=np.number
    ).columns
)

print(f"Numerical Features : {len(numerical_columns)}")

display(
    df[numerical_columns]
    .describe()
    .T
    .round(2)
)
```

### Section 10: Categorical Feature Audit


```python
# Categorical Feature Analysis

categorical_columns = (
    df.select_dtypes(
        include="str"
    ).columns
)

print('-' * 24)
print(f"Categorical Features : {len(categorical_columns)}")
print('-' * 24)

display(categorical_columns.tolist())

print('-' * 21)

# Unique Value Analysis

print("Unique Value Analysis")
print('-' * 21)

for col in categorical_columns:
    print(
        f"{col}: {df[col].nunique()}"
    )
```

```python
print("\n" + "-" * 32)
print("Categorical Cardinality Analysis")
print("-" * 32)

cardinality_report = pd.DataFrame({
    "Column": categorical_columns,
    "Unique Values": [
        df[column].nunique()
        for column in categorical_columns
    ]
})

display(
    cardinality_report.sort_values(
        by="Unique Values",
        ascending=False
    )
)
```

```python
# Numerical Validation Report
print("Numerical Validation Report")

numerical_columns = (
    df.select_dtypes(
        include=np.number
        ).columns
)

validation_report = []

for column in numerical_columns:

    validation_report.append({

        "Column"            :   column,
        "Minimum Value"     :   df[column].min(),
        "Maximum Value"     :   df[column].max(),
        "Negative Values"   :   (df[column] < 0).sum()

    })

validation_report_pd = pd.DataFrame(
    validation_report
)

display(validation_report_pd)
```

## Final Audit Summary


```python
audit_summary = pd.DataFrame({

    "Metric": [
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicate Rows",
        "Numerical Features",
        "Categorical Features"
    ],

    "Value": [
        len(df),
        len(df.columns),
        df.isnull().sum().sum(),
        df.duplicated().sum(),
        len(numerical_columns),
        len(categorical_columns)
    ]
})

print("\n" + "-" * 27)
print("Final Dataset Audit Summary")
print("-" * 27)

display(audit_summary)
```
