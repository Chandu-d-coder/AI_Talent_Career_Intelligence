### Section 1: Imports

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("\nEnvironment Information")
print("-" * 22)

print(f"Pandas Version : {pd.__version__}")
print(f"NumPy Version  : {np.__version__}")
```

### Section 2: Load Dataset and copy

```python
df = pd.read_excel("../data/raw/ai_jobs.xlsx")

df_clean = df.copy()

print("-" * 28)
print("Dataset loaded successfully")
print("-" * 28)

print(f"Rows    : {df_clean.shape[0]:,}")
print(f"Columns : {df_clean.shape[1]}")

df_clean.head()
```

### Section 3: Missing value treatment

```python
numerical_columns = (
    df_clean.select_dtypes(
        include=np.number
    ).columns
)

for col in numerical_columns:

    df_clean[col] = (
        df_clean[col]
        .fillna(
            df_clean[col].median()
        )
    )

df_clean.head()
```

### Section 4: Categorical Columns treatment

```python
categorical_columns = (
    df_clean.select_dtypes(
        include=["object"]
    ).columns
)

for col in categorical_columns:

    df_clean[col] = (
        df_clean[col]
        .fillna(
            df_clean[col].mode()[0]
        )
    )

df_clean.head()
```

### Section 5: Remove duplicate rows

```python
before_rows = len(df_clean)

df_clean = (
    df_clean
    .drop_duplicates()
)

after_rows = len(df_clean)

print(
    f"\nRemoved {before_rows - after_rows} duplicate rows"
)
```

### Section 6: Remove Leading and Trailing Spaces

```python
for col in categorical_columns:

    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.strip()
    )

print("\nRemoved Leading and Trailing Spaces Successfully")
```

### Section 7: Text Standardization

```python
print("-" * 20)
print("Text Standardization")
print("-" * 20)

categorical_columns = (
    df_clean.select_dtypes(
        include="object"
    ).columns
)

# Country mapping
df_clean["country"] = (
    df_clean["country"]
    .astype(str)
    .str.strip()
    .str.title()
)

country_mapping = {
    "Uae" : "UAE",
    "Usa" : "USA",
    "Uk" : "UK"
    }

df_clean["country"] = (
    df_clean["country"]
    .replace(country_mapping)
)

# Job role mapping
df_clean["job_role"] = (
    df_clean["job_role"]
    .astype(str)
    .str.strip()
    .str.title()
)

job_role_mapping = {
    "Ai Engineer" : "AI Engineer",
    "Software Engineer Ai" : "AI Software Engineer",
    "Nlp Engineer" : "NLP Engineer"
    }

df_clean["job_role"] = (
    df_clean["job_role"]
    .replace(job_role_mapping)
)

# AI Specialization mapping
df_clean["ai_specialization"] = (
    df_clean["ai_specialization"]
    .astype(str)
    .str.strip()
    .str.title()
)

ai_specialization_mapping = {
    "Llm" : "LLM",
    "Nlp" : "NLP",
    "Mlops" : "MLOps",
    "Generative Ai" : "Generative AI"
    }

df_clean["ai_specialization"] = (
    df_clean["ai_specialization"]
    .replace(ai_specialization_mapping)
)


# Round Score
df_clean["hiring_difficulty_score"] = (
    df_clean["hiring_difficulty_score"]
    .round(2)
)

# Standardize Remaining Text Columns
for col in categorical_columns:

    if col == "country" or col == "job_role" or col == "ai_specialization":
        continue

    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.strip()
        .str.title()
    )

print("Text Standardization Completed")

# Check modifications

job = df["job_role"].unique().tolist()
ai_spec = df["ai_specialization"].unique().tolist()
country = df["country"].unique().tolist()

print(job)
print(ai_spec)
print(country)
```

### Section 8: Replace Empty Strings

```python
for col in categorical_columns:

    df_clean[col] = (
        df_clean[col]
        .replace(
            "",
            np.nan
        )
    )

print("\nReplaced Empty Strings Successfully")
```

### Section 9: Replace index

```python
# Reset the index

df_clean = (
    df_clean
    .reset_index(
        drop=True
    )
)
```

### Section 10: Export Clean Dataset

```python
import time

print("-" * 25)
print("Clean Dataset Export")
print("-" * 25)

steps = [
    "Preparing dataset...",
    "Validating records...",
    "Writing CSV file...",
    "Finalizing export..."
]

for step in steps:
    print(step)
    time.sleep(0.5)

df_clean.to_csv(
    "../data/processed/cleaned_ai_jobs.csv",
    index=False
)

print()
print("Export Completed Successfully")
print("-" * 50)

print(
    "Saved To: ../data/processed/cleaned_ai_jobs.csv"
)

print(
    f"Final Shape : {df_clean.shape}"
)
```
