## Phase 3: Intelligence Framework Design

### Objective
Design the analytical blueprint for the AI Career Intelligence Platform before performing any analysis.

This phase defines:
- Intelligence opportunities
- Business questions
- Analytical targets
- Driver variables
- KPIs
- Hypotheses
- Analysis priorities


## Mission 1: Intelligence Opportunity Discovery

### 1. Salary Intelligence
- Does country affect salary?
- Does experience affect salary?
- Does hiring difficulty affect salary?
- Does number of interview rounds affect salary?
- Does industry affect salary?

### 2. Career Growth Intelligence
- Does skill demand improve career growth?
- What affects career growth most?
- Does AI specialization impact promotion speed?
- Does experience influence promotion speed?

### 3. Workforce Risk Intelligence
- Does AI adoption affect job security?
- Does layoff risk vary by country?
- Does company behavior affect job security?
- Which jobs are safest?

### 4. Talent Market Intelligence
- What affects job openings?
- Which factors increase hiring demand?
- Why are some skills more demanded?

### 5. Employee Experience Intelligence
- Does work mode affect work-life balance?
- Does company rating affect satisfaction?
- What improves employee happiness?

### 6. Recruitment Intelligence
- Why do candidates reject offers?
- Is salary the main reason?
- Does hiring difficulty impact acceptance?

## Mission 4: Hypothesis Registry

H1. Higher experience leads to higher salary.

H2. Higher skill demand scores lead to higher salaries.

H3. Certain AI specializations command significantly higher salaries than others.

H4. Salary varies across countries and industries.

H5. Higher hiring difficulty is associated with higher salaries.

H6. Higher skill demand improves career growth.

H7. More experienced professionals achieve faster promotions.

H8. Higher AI adoption influences job security.

H9. Higher automation risk reduces job security.

H10. Lower layoff risk is associated with better promotion opportunities.

H11. More vacation days improve work-life balance.

H12. Higher weekly working hours reduce work-life balance.

H13. Higher weekly working hours reduce employee satisfaction.

H14. Higher salaries increase offer acceptance rates.

H15. More interview rounds reduce offer acceptance rates.

H16. Higher hiring difficulty impacts offer acceptance rates.

H17. Highly demanded skills create more job openings.

H18. Higher company ratings improve employee satisfaction.

H19. Remote work improves employee satisfaction.

H20. Remote work improves work-life balance.
